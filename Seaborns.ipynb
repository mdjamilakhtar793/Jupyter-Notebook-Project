```python
# Heatmaps
flight_df = sbn.load_dataset('flights')
flight_df.head()
```

```python
flight_df = flight_df.pivot(index="month", columns="year", values="passengers")
flight_df.head()
```

```python

plt.figure(figsize=(12,6))
ax = sbn.heatmap(flight_df)
ax
     
```

```python
plt.figure(figsize=(14,8))
ax = sbn.heatmap(flight_df, annot = True, fmt = 'd')
ax
     
```

```python
plt.figure(figsize=(14,8))
ax = sbn.heatmap(flight_df, annot = True, fmt = 'd' , linecolor='k', linewidths='5')
ax
```

```python
plt.figure(figsize=(14,8))
ax = sbn.heatmap(flight_df, annot = True, fmt = 'd',
                  linecolor='k', linewidths='5',
                  cmap = 'Blues')
ax
```

```python
plt.figure(figsize=(14,8))
sbn.heatmap(flight_df, cbar=False)
     
```

```python
grid_kws = {"height_ratios": (.3, .05), "hspace": .4}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sbn.heatmap(flight_df, cbar_kws={"orientation": "horizontal"} , ax=ax,
                 cbar_ax=cbar_ax,)
```
