```python
# arr.reshape((rows,cols)) it devide into 2 and so on ....
n=np.arange(1,100,2)
n

n=n.reshape((10,5))
n

# flatten() : it return copy of Orginal Arrays, if we modify any value of orginal array is not effected.
# it is slower than ravel as it occupies memory and it is method of NdArrays.
n=n.flatten()
n
# ravel() : it return only referance/view of orginal array , if we modify the array we noticed that the valueof Orginal arrays also changes,
# it is faster than flatten() as it does not occupy any memory.
# it is a library level function
n=n.ravel()
n

#  
```

```python
# Mathematical Operations 
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")

a=np.arange(0,18).reshape(6,3)
b=np.arange(20,38).reshape(6,3)
print(a)
print(b)

# a+b
np.add(a,b)
np.subtract(a,b)
np.subtract(a,b)
np.multiply(a,b)
np.divide(a,b)

# Matrix Maltilpication
print(a.shape)
print(b.shape)

b=b.reshape((3,6))
a@b
a.dot(b)
a.max()
b.argmax()
b.min()
np.sum(b)
np.sum(b,axis=1)
np.sum(b,axis=0)
np.mean(b)
np.sqrt(a)
np.std(b)
np.log(b)
```

```python
#  Trigonometric Operation

x=np.arange(1,11)
y=np.arange(10,110,10)

plt.figure(figsize=(6,6))
plt.plot(x,y,"r--")
plt.show()
```

```python
# Plotting trigonomateric Curves
xx_sin=np.arange(0,2*np.pi,0.1)
yy_sin=np.sin(xx_sin)
print(yy_sin)

plt.figure(figsize=(6,6))
plt.plot(xx_sin,yy_sin)
plt.title('Sin Curve')
plt.show()
```

```python
xx_cos=np.arange(0,2*np.pi,0.1)
yy_cos=np.cos(xx_cos)
print(yy_cos)

plt.figure(figsize=(6,6))
plt.plot(xx_cos,yy_cos)
plt.title('Cos Curve')
plt.show()
```

```python
xx_tan=np.arange(0,2*np.pi,0.1)
yy_tan=np.tan(xx_tan)
print(yy_tan)

plt.figure(figsize=(6,6))
plt.plot(xx_tan,yy_tan)
plt.title('Tan Curve')
plt.show()
```

```python
# Subplot in single Paper

plt.figure(figsize=(6,6))
plt.subplot(2,2,1)
plt.plot(xx_sin,yy_sin,"r-")
plt.title('Sin Curve')

plt.subplot(2,2,2) # Size 
plt.plot(xx_cos,yy_cos,"b-")
plt.title('Cos Curve')

plt.subplot(2,2,3)
plt.plot(xx_tan,yy_tan,"g-")
plt.title('Tan Curve')

plt.subplot(2,2,4)
plt.plot(xx_cos,yy_cos,"w-")
plt.title('Cot Curve')
plt.show()
```
