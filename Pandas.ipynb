```python
# Merge Function with Data Frame
df1=pd.DataFrame({"Roll No.":[1,2,3,4,5],'Physics':[34,37,34,89,12]})
df1
```

```python
df2=pd.DataFrame({"Roll No.":[1,2,3,4,5],'Chemistry':[34,37,34,89,12]})
df2
```

```python
pd.merge(df1,df2, on = "Roll No.")
```

```python
pd.merge(df1,df2) # on by default intersecting column is token
```

```python
df3=pd.DataFrame({"Roll No.":[1,2,3,6,7],'Physics':[34,37,34,89,12]})
df4=pd.DataFrame({"Roll No.":[1,2,3,8,9],'Chemistry':[34,37,34,89,12]})

pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,how='left') 
```

```python
pd.merge(df3,df4,how='outer') 
```
